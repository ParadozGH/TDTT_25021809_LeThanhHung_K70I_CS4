```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def volume(self):
        return 3.14 * (self.radius ** 2) * self.height
    def surface_area(self):
        return 2 * 3.14 * self.radius * (self.radius + self.height)
radius, height = map(int, input().split())
cyl = Cylinder(radius, height)
print(f"{cyl.volume():.2f} {cyl.surface_area():.2f}")
```

```python
#W8A2
class Date:
    def __init__(self, date, month, year):
        self.date = date
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, s):
        lst = s.split('/')
        date = lst[0]
        month = lst[1]
        year = lst[2]
        return cls(date, month, year)
    
    def check_leap(self):
        year_ = int(self.year)
        if year_ % 400 == 0 or (year_ % 4 == 0 and year_ % 100 != 0):
            return True
        else:
            return False

    def next_day(self):
        day_31 = [1,3,5,7,8,10,12]
        day_30 = [4,6,9,11]
        if int(self.month) in day_31:
            if self.date == '31':
                if self.month == '12':
                    self.date = '01'
                    self.month = '01'
                    if int(self.year) <= 998:
                        self.year = '0' + str(int(self.year) + 1)
                    else:
                        self.year = str(int(self.year) + 1)
                else:
                    self.date = '01'
                    if int(self.month) <= 8:
                        self.month = '0' + str(int(self.month) + 1)
                    else:
                        self.month = str(int(self.month) + 1)
            else:
                if int(self.date) <= 8:
                    self.date = '0' + str(int(self.date) + 1)
                else:
                    self.date = str(int(self.date) + 1)
        elif int(self.month) in day_30:
            if self.date == '30':
                self.date = '01'
                if int(self.month) <= 8:
                    self.month = '0' + str(int(self.month) + 1)
                else:
                    self.month = str(int(self.month) + 1)
            else:
                if int(self.date) <= 8:
                    self.date = '0' + str(int(self.date) + 1)
                else:
                    self.date = str(int(self.date) + 1)
        elif self.month == '02':
            if self.check_leap:
                if self.date == '28':
                    self.date = '29'
                elif self.date == '29':
                    self.date = '01'
                    self.month = '03'
                else:
                    if int(self.date) <= 8:
                        self.date = '0' + str(int(self.date) + 1)
                    else:
                        self.date = str(int(self.date) + 1)
            else:
                if self.date == '28':
                    self.date = '01'
                    self.month = '03'
                else:
                    if int(self.date) <= 8:
                        self.date = '0' + str(int(self.date) + 1)
                    else:
                        self.date = str(int(self.date) + 1)
        return f"{self.date}/{self.month}/{self.year}"
s = input()
time = Date.from_string(s)
print(time.next_day())
                

                
                
```

```python
#W8A3
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5
    def is_origin(self):
        return self.x == 0 and self.y == 0
    def is_on_x_axis(self):
        return self.y == 0
    def is_on_y_axis(self):
        return self.x == 0
p = Point2D(*map(int, input().split()))
print(f" {p.is_origin()} {p.is_on_x_axis()} {p.is_on_y_axis()} {p.distance_from_origin():.2f}")
```

```python
#W8A4
class Calculator:
    def __init__ (self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b  
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Cannot divide by zero"
    def modulus(self):
        if self.b != 0:
            return self.a % self.b
        else:
            return "Cannot perform modulus by zero"
    def power(self):
        return self.a ** self.b
    def set_numbers(self, a, b):
        self.a = a
        self.b = b
calc = Calculator(*map(int, input().split()))
while True:
    print("Add:", calc.add())
    print("Subtract:", calc.subtract())
    print("Product:", calc.product())
    print("Divide:", calc.divide())
    print("Modulus:", calc.modulus())
    print("Power:", calc.power())
    print("Bạn có muốn thoát không ?")
    operation = input()
    if operation.lower() == "yes":
        break
    elif operation == "set":
        calc.set_numbers(*map(int, input().split()))

```

```python
#W8A5
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price  = price
class ShoppingCart:
    def __init__(self, cart: list):
        self.cart = cart
    def empty_cart(self):
        self.cart = []
    def add_items(self, product: Product):
        self.cart.append(product)
    def delete_items(self, name):
        for items in self.cart:
            if items.name == name:
                self.cart.remove(items)
    def check_empty(self):
        if self.cart == []:
            return "Empty"
        else:
            return f"Has {len(self.cart)} item(s)"
    def sum_price(self):
        sum = 0
        for items in self.cart:
            sum += items.price
        return sum
    def display_cart(self):
        if self.cart == []:
            print("Empty")
        else:
            for items in self.cart:
                print(f"                 {items.name}         {items.price}")
    def clear_cart(self):
        self.cart = []
print("""                ======= MENU =======
        1. Khởi tạo với danh sách rỗng
        2. Thêm vào giỏ hàng một sản phẩm mới
        3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)
        4. Kiểm tra xem giỏ hàng có rỗng không
        5. Tính tổng tiền giỏ hàng
        6. Hiển thị toàn bộ sản phẩm trong giỏ
        7. Xóa toàn bộ giỏ hàng """)
Cart = ShoppingCart([])
while True:
    print( "Nhập truy vấn:", end = " ")
    s = int(input())
    print("\n")
    if s == 1:
        Cart.empty_cart()
    elif s == 2:
        print("Nhập tên sản phẩm:", end = " ")
        name = input()
        print("\nNhập giá sản phẩm:", end = " ")
        price = int(input())
        product = Product(name,price)
        Cart.add_items(product)
    elif s == 3:
        print("Nhập tên sản phẩm cần xoá:", end = " ")
        name = input()
        Cart.delete_items(name)
        print("\nĐã xoá sản phẩm !")
    elif s == 4:
        print(Cart.check_empty())
    elif s == 5:
        print("Tổng số tiền là:",Cart.sum_price())
    elif s == 6:
        print("""                Danh sách vật phẩm:
                 Tên ======= Giá""")
        Cart.display_cart()
    elif s == 7:
        Cart.clear_cart()
        print("Đã xoá giỏ hàng !")
    else:
        print("Nhập lại truy vấn: ", end = " ")
        s = input()
    print("\nCó muốn thoát khỏi chương trình không ?", end = " ")
    ans = input()
    if ans.lower() == "yes":
        break
    else:
        print("\nTiếp tục chương trình")




    
                    
    
```

```python
#W8A6
class Rectangle:

    def __init__(self, length, width,factor):
        self.length = length
        self.width = width
        self.factor=factor
    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)
    def scale(self):
        self.length *= self.factor
        self.width *= self.factor
h = Rectangle(*map(int, input().split()))
h.scale()
print(f"{h.area():.2f} {h.perimeter():.2f}")
```

```python
#W8A7
import math
from math import gcd
class Fraction:
    def __init__(self,num,op,den,num2,den2):
        self.num = num
        self.den = den
        self.op = op
        self.num2 = num2
        self.den2 = den2
    @classmethod
    def from_string(cls, frac_str):
        lst = frac_str.split()
        num = int(lst[0])
        den = int(lst[1])
        op = lst[2]
        num2 = int(lst[3])
        den2 = int(lst[4])
        return cls(num,op,den,num2,den2)
    def operation(self):
        if self.op == '+':
            num_res = self.num * self.den2 + self.num2 * self.den
            den_res = self.den * self.den2
        elif self.op == '-':
            num_res = self.num * self.den2 - self.num2 * self.den
            den_res = self.den * self.den2
        elif self.op == '*':
            num_res = self.num * self.num2
            den_res = self.den * self.den2
        elif self.op == '/':
            num_res = self.num * self.den2
            den_res = self.den * self.num2
        else:
            return "Invalid operator"
        common = gcd(abs(num_res), abs(den_res))
        num_res //= common
        den_res //= common
        if den_res < 0:
            num_res = -num_res
            den_res = -den_res
        return f"{num_res}/{den_res}"
frac_str = input()
fraction = Fraction.from_string(frac_str)
print(fraction.operation())
            
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        if amount > self.balance:
            pass
        else:
            self.balance -= amount
    def __str__(self):
        return f"{self.balance:.2f}"
owner, balance = input().split()
account = BankAccount(owner, float(balance))
n = int(input())
for _ in range(n):
    action, amount = input().split()
    amount = float(amount)
    if action == "DEPOSIT":
        account.deposit(amount)
    elif action == "WITHDRAW":
        account.withdraw(amount)
print(account.__str__())
```

```python
#W8A9
class Student:
    seen = {}
    def __init__(self, name, score):
        self.name = name
        self.score = score
        self.cnt = 0
    def add_score(self, subject, score):
        self.score += score
        if subject not in Student.seen:
            Student.seen[subject] = True
            self.cnt += 1
    @property
    def avg_score(self):
        return f"{self.score / self.cnt:.2f}"
    def rank(self):
        if float(self.avg_score) >= 8:
            return "Excellent"
        elif float(self.avg_score) >= 6.5 and float(self.avg_score) < 8:
            return "Good"
        elif float(self.avg_score) >= 5 and float(self.avg_score) < 6.5:
            return "Average"
        else:
            return "Poor"
student = Student(input(), 0)
n = int(input())
for _ in range(n):
    subject, score = input().split()
    score = float(score)
    student.add_score(subject, score)
print(student.name,student.avg_score, student.rank())
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.books = []
    @classmethod
    def from_string(cls, book_str):
        lst = book_str.split(';')
        title = lst[0]
        author = lst[1]
        year = int(lst[2])
        return Book(title, author, year)
    def add_book(self, info):
        book = Library.from_string(info)
        self.books.append(book)
    def count_by_author(self, author_name):
        cnt = 0
        for book in self.books:
            if book.author == author_name:
                cnt += 1
        return cnt
    def find_by_year(self, year):
        cnt = 0
        for book in self.books:
            if book.year == year:
                cnt += 1
        return cnt
n = int(input())
library = Library()
for _ in range(n):
    info = input().split()
    if info[0] == "ADD":
        library.add_book(info[1])
    elif info[0] == "COUNT":
        author_name = info[1]
        print(library.count_by_author(author_name))
    elif info[0] == "COUNT_YEAR":
        year = int(info[1])
        print(library.find_by_year(year))
```

```python
#W8A11
import math
from math import acos
from math import sqrt
from math import degrees
class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def vector_(self, other):
        return Point(other.x - self.x, other.y - self.y, other.z - self.z)
    def cross_product(self, other):
        cross_x = self.y * other.z - self.z * other.y
        cross_y = self.z * other.x - self.x * other.z
        cross_z = self.x * other.y - self.y * other.x
        return Point(cross_x, cross_y, cross_z)
    def dot_product(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def magnitude(self):
        return sqrt(self.x**2 + self.y**2 + self.z**2)
    def angle_between(self, other):
        cos_angle = self.dot_product(other) / (self.magnitude() * other.magnitude())
        angle_rad = acos(cos_angle)
        return degrees(angle_rad)
p1 = Point(*map(float, input().split()))
p2 = Point(*map(float, input().split()))
p3 = Point(*map(float, input().split()))
p4 = Point(*map(float, input().split()))
vector_AB = p1.vector_(p2)
vector_AC = p1.vector_(p3)
vector_CD = p3.vector_(p4)
vector_CB = p3.vector_(p2)
n1 = vector_AB.cross_product(vector_AC)
n2 = vector_CD.cross_product(vector_CB)
angle = n1.angle_between(n2)
print(f"{angle:.2f}")

```

```python
#W8A13
import math
from math import sqrt
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    @property
    def check(self):
        if self.edge1 + self.edge2 > self.edge3 and self.edge2 + self.edge3 > self.edge1 and self.edge1 + self.edge3 > self.edge2:
            return True
        else:
            return False
    def get_area(self):
        p = (self.edge1 + self.edge2 + self.edge3)/2
        s = sqrt(p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))
        return f"{s:.2f}"
tri = Triangle(*map(int,input().split()))
if tri.check:
    print(tri.get_area())
else:
    print("Invalid")
```
