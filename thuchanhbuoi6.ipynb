```python
#W6A4
input_string = input()
pairs = input_string.split()
result_dict = {}

for pair in pairs:
    key, value = pair.split(':', 1) 

    if key in result_dict:
        result_dict[key].append(value)
    else:
        result_dict[key] = [value]

print(result_dict)
```

```python
#W6A5
a = list(map(int,input().split()))
freq ={'positive' : 0,'negative': 0,'zeros': 0}
for i in range(len(a)):
    if a[i] == 0:
        freq['zeros'] += 1
    elif a[i] > 0:
        freq['positive'] += 1
    else:
        freq['negative'] += 1
print(freq)
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())
if d1 == {} or d2 == {}:
    print({})
else:
    d3 = {}
    for key in d1:
        if key in d2:
            d3[key] = d1[key] + d2[key]
        else:
            d3[key] = d1[key]
    for key in d2:
        if key not in d3:
            d3[key] = d2[key]
    d4 = dict(sorted(d3.items()))
    print(d4)
```

```python
#W6A10
lst = list(map(int, input().split()))
k = int(input())

seen = set()
ket_qua_set = set()

for x in lst:
    target = k - x
    if target in seen:
        ket_qua_set.add((target, x))
    seen.add(x)
print(sorted(list(ket_qua_set)))
```

```python
#W6A12
a = list(map(int,input().split()))
a.sort()
freq = {}
for i in range(len(a)):
    if a[i] not in freq:
        freq[a[i]] = 1
    else:
        freq[a[i]] += 1
max_freq = max(freq.values())
for x in freq:
    if freq[x] == max_freq:
        print(x)
        break
```

```python
#W6A18
n,m,k = map(int,input().split())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
sumrowk = 0
sumcolk = 0
for i in range(m):
    sumrowk += matrix[k-1][i]
for i in range(n):
    sumcolk += matrix[i][k-1]
print(sumrowk,sumcolk)
```
